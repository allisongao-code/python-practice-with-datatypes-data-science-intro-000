### Q3

Provide a simplified table that only lists the most common request submission method for each complaint type. There should be one row per complaint type, capturing the most common submission method for that complaint type.

```python
counts = df.groupby('Open Data Channel Type' 'Complaint Type').size().reset_index(name='count')
valid_open_types = counts['Open Data Channel Type'][counts['count'] > 4]
df_cleaned = df[df['Open Data Channel Type'].isin(valid_open_types)]
df_cleaned.groupby(['Open Data Channel Type', 'Complaint Type']).size().nlargest(20).reset_index(name='count')
```
